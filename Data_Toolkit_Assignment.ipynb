```python
'''
Q3) Explain the following terms:
    The difference in np.array, np.asarray and np.asanyarrayX
    The difference between Deep copy and shallow copy.

Ans) 1. Difference between np.array, np.asarray, and np.asanyarray:
np.array: Creates a new NumPy array and copies the data by default. If the input is already a NumPy array, it still creates a copy unless specified otherwise.
np.asarray: Converts the input to a NumPy array but avoids copying if the input is already a NumPy array. It’s more memory efficient.
np.asanyarray: Similar to np.asarray, but allows for the preservation of subclasses of ndarray like masked arrays, rather than forcing a conversion to a base ndarray.

2. Difference between Deep Copy and Shallow Copy:
Shallow Copy: Creates a new object, but only references the original inner objects. Changes to the mutable elements inside the new object affect the original.
Deep Copy: Creates a completely independent copy of the object and its inner elements. Changes to the new object do not affect the original.
'''
```

```python
'''
Q5) Create a NumPy array with random integers between 1 and 10 of shape (5, 6). After creating the array
perform the following operations:

 a)Extract all even integers from array.

 b)Extract all odd integers from array.  '''

import numpy as np

# Create a NumPy array with random integers between 1 and 10 of shape (5, 6)
array = np.random.randint(1, 11, (5, 6))

# a) Extract all even integers from the array
even_integers = array[array % 2 == 0]

# b) Extract all odd integers from the array
odd_integers = array[array % 2 != 0]

print("Original Array:\n", array)
print("Even Integers:\n", even_integers)
print("Odd Integers:\n", odd_integers)
```

```python
'''
Q6) Create a 3D NumPy array of shape (3,3,3) containing random integers between 1 and 10. Perform the  following operations:

 a) Find the indices of the maximum values along each depth level (third axis).

 b) Perform element-wise multiplication of between both array.  '''

import numpy as np

# Create a 3D NumPy array of shape (3, 3, 3) containing random integers between 1 and 10
array = np.random.randint(1, 11, (3, 3, 3))

# a) Find the indices of the maximum values along each depth level (third axis)
max_indices = np.argmax(array, axis=2)

# b) Perform element-wise multiplication of the array with itself
multiplied_array = np.multiply(array, array)

print("Original 3D Array:\n", array)
print("Indices of Maximum Values Along Each Depth Level:\n", max_indices)
print("Element-wise Multiplication of Array with Itself:\n", multiplied_array)
```

```python
'''
Q14) Create a time-series dataset in a Pandas DataFrame with columns: 'Date', 'Temperature', 'Humidity' and  Perform the following tasks using Matplotlib:


a) Plot the 'Temperature' and 'Humidity' on the same plot with different y-axes (left y-axis for 'Temperature' and  right y-axis for 'Humidity').

b) Label the x-axis as 'Date'.

c) Set the title of the plot as 'Temperature and Humidity Over Time'.   '''

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Create a time-series dataset
date_range = pd.date_range(start='2024-01-01', periods=30, freq='D')  # 30 days of data
temperature = np.random.uniform(15, 35, 30)  # Random temperatures between 15 and 35 degrees
humidity = np.random.uniform(40, 90, 30)    # Random humidity percentages between 40% and 90%

# Create a DataFrame
df = pd.DataFrame({
    'Date': date_range,
    'Temperature': temperature,
    'Humidity': humidity
})

# a) Plot 'Temperature' and 'Humidity' on the same plot with different y-axes
fig, ax1 = plt.subplots()

# Left y-axis for 'Temperature'
ax1.plot(df['Date'], df['Temperature'], color='red', label='Temperature')
ax1.set_ylabel('Temperature (°C)', color='red')
ax1.tick_params(axis='y', labelcolor='red')

# Right y-axis for 'Humidity'
ax2 = ax1.twinx()
ax2.plot(df['Date'], df['Humidity'], color='blue', label='Humidity')
ax2.set_ylabel('Humidity (%)', color='blue')
ax2.tick_params(axis='y', labelcolor='blue')

# b) Label the x-axis as 'Date'
ax1.set_xlabel('Date')

# c) Set the title of the plot
plt.title('Temperature and Humidity Over Time')

# Show the plot
plt.show()


```

```python
'''
Q15) Create a NumPy array data containing 1000 samples from a normal distribution. Perform the following tasks using Matplotlib:
a) Plot a histogram of the data with 30 bins.
b) Overlay a line plot representing the normal distribution's probability density function (PDF).
c) Label the x-axis as 'Value' and the y-axis as 'Frequency/Probability'.
d) Set the title of the plot as 'Histogram with PDF Overlay'.   '''

import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# a) Create a NumPy array containing 1000 samples from a normal distribution
data = np.random.normal(loc=0, scale=1, size=1000)

# Create a histogram of the data with 30 bins
plt.figure(figsize=(10, 6))
count, bins, ignored = plt.hist(data, bins=30, density=True, alpha=0.6, color='blue', edgecolor='black')

# b) Overlay a line plot representing the normal distribution's probability density function (PDF)
xmin, xmax = plt.xlim()
x = np.linspace(xmin, xmax, 100)
p = norm.pdf(x, np.mean(data), np.std(data))
plt.plot(x, p, 'r', linewidth=2)

# c) Label the x-axis as 'Value' and the y-axis as 'Frequency/Probability'
plt.xlabel('Value')
plt.ylabel('Frequency/Probability')

# d) Set the title of the plot as 'Histogram with PDF Overlay'
plt.title('Histogram with PDF Overlay')

# Show the plot
plt.grid(True)
plt.show()
```

```python
'''
Q17) Create a Seaborn scatter plot of two random arrays, color points based on their position relative to the
origin (quadrants), add a legend, label the axes, and set the title as 'Quadrant-wise Scatter Plot'.   '''

import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Create two random arrays
x = np.random.uniform(-10, 10, 100)  # Random x values between -10 and 10
y = np.random.uniform(-10, 10, 100)  # Random y values between -10 and 10

# Create a DataFrame to hold the data
df = pd.DataFrame({'X': x, 'Y': y})

# Determine the quadrant for each point
df['Quadrant'] = np.where(
    (df['X'] > 0) & (df['Y'] > 0), 'Quadrant 1',
    np.where(
        (df['X'] < 0) & (df['Y'] > 0), 'Quadrant 2',
        np.where(
            (df['X'] < 0) & (df['Y'] < 0), 'Quadrant 3',
            'Quadrant 4'
        )
    )
)

# Create a Seaborn scatter plot
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df, x='X', y='Y', hue='Quadrant', palette='Set1', style='Quadrant')

# Label the axes
plt.xlabel('X-axis')
plt.ylabel('Y-axis')

# Set the title
plt.title('Quadrant-wise Scatter Plot')

# Add a legend
plt.legend(title='Quadrants')

# Show the plot
plt.grid(True)
plt.show()


```

```python
'''
Q18) With Bokeh, plot a line chart of a sine wave function, add grid lines, label the axes, and set the title as 'Sine Wave Function.   '''

import numpy as np
import matplotlib.pyplot as plt

# Generate x values from 0 to 2π
x = np.linspace(0, 2 * np.pi, 100)

# Calculate the sine of x
y = np.sin(x)

# Create a line chart of the sine wave function
plt.figure(figsize=(10, 6))
plt.plot(x, y, label='y = sin(x)', color='blue')

# Add grid lines
plt.grid(True)

# Label the axes
plt.xlabel('X values (radians)')
plt.ylabel('Y values')

# Set the title
plt.title('Sine Wave Function')

# Add a legend
plt.legend()

# Show the plot
plt.show()




```

```python
'''
Q19) Using Bokeh, generate a bar chart of randomly generated categorical data, color bars based on their
values, add hover tooltips to display exact values, label the axes, and set the title as 'Random Categorical Bar Chart'.   '''

# Install Bokeh if not already installed
!pip install bokeh

from bokeh.io import output_notebook, show
from bokeh.plotting import figure
from bokeh.models import ColumnDataSource, HoverTool
import numpy as np
import pandas as pd

# Enable output for Jupyter-like environments
output_notebook()

# Generate random categorical data
categories = [f'Category {i}' for i in range(1, 11)]  # 10 categories
values = np.random.randint(10, 100, size=len(categories))  # Random values for each category

# Create a DataFrame
data = pd.DataFrame({'Category': categories, 'Values': values})

# Create a ColumnDataSource
source = ColumnDataSource(data=data)

# Create a figure
p = figure(x_range=categories, title='Random Categorical Bar Chart',
           toolbar_location=None, tools="")

# Create a bar chart
p.vbar(x='Category', top='Values', source=source,
        line_color='white', fill_color='navy')

# Add hover tooltips
hover = HoverTool()
hover.tooltips = [("Category", "@Category"), ("Value", "@Values")]
p.add_tools(hover)

# Label the axes
p.xaxis.axis_label = 'Categories'
p.yaxis.axis_label = 'Values'
p.xaxis.major_label_orientation = "vertical"

# Show the plot
show(p)

```
